```python
ax=sns.lmplot(x='displ',y='hwy',hue='cyl',data=df)
ax.set(xlabel='displ',)
plt.xlabel('displ',fontdict=dict(fontsize=15))
```

```python
df.cyl.unique()
# hue -groupby 
df.info()
df.year.unique()
```

```python
ax=sns.lmplot(x='displ',y='hwy',hue='year',data=df)
ax.set(xlabel='displ',)
plt.xlabel('displ',fontdict=dict(fontsize=15))
```

```python
ax=sns.lmplot(x='displ',y='hwy',hue='year',data=df,fit_reg=False,markers=['o','x'],scatter=False)
ax.set(xlabel='displ',)
ax.set_xlabels(fontsize=15,fontdict=dict(color='maroon'))
# plt.xlabel('displ',fontdict=dict(fontsize=15),)
```

```python
ax=sns.lmplot(x='displ',y='hwy',hue='year',data=df,fit_reg=True,markers=['o','x'],scatter=True,
             ci=0,palette='twilight',)
sns.color_palette(palette= 'rainbow',n_colors=5)
ax.set(xlabel='displ',)
ax.set_xlabels(fontsize=15,fontdict=dict(color='maroon'))
ax.set_ylabels(fontsize=15,rotation=70)
ax.set_yticklabels(rotation=75,fontdict=dict(fontsize=10,color='maroon'))
```
