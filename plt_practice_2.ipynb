```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(20,70,10,endpoint=False,dtype=int)
y=np.random.randint(low=1,high=10,size=10,dtype=int)
plt.figure(figsize=(10,4),facecolor='#807a8b',edgecolor='red',dpi=100)
plt.plot(x,y,'*:m',lw=2,mec='k',mew=2,mfc='pink',)
```

```python
plt.figure(figsize=(10,4),facecolor='#807a8b',edgecolor='red',dpi=100)
plt.plot(x,y,'*:m',lw=2,mec='k',mew=2,mfc='pink',)
plt.annotate('Local Maxima',xy=(55,6),xytext=(57,5.5),arrowprops=dict(
facecolor='green',shrink=0.05),)
```

```python
plt.figure(figsize=(10,4),facecolor='#807a8b',edgecolor='red',dpi=100)
plt.plot(x,y,'*:m',lw=2,mec='k',mew=2,mfc='pink',)
plt.annotate('Local Maxima',xy=(55,6),xytext=(55,3),arrowprops=dict(
facecolor='green',shrink=0.005,width=1,),xycoords='data',textcoords='offset pixels')
```

```python
plt.figure(figsize=(10,4),facecolor='#807a8b',edgecolor='red',dpi=100)
plt.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
plt.annotate('Local Maxima',xy=(55,6),xytext=(55,3),arrowprops=dict(
facecolor='green',arrowstyle='simple',),xycoords='data',textcoords='offset pixels')
```

```python
plt.figure(figsize=(15,5),facecolor='#807a8b',edgecolor='red',dpi=250)

plt.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)

bbox = dict(boxstyle ="round", fc ="0.8")
arrowprops = dict(
    arrowstyle = "<->",
    connectionstyle = "angle, angleA = 0, angleB = 90,\
    rad = 10")
xdata,ydata,offset=55,6,-20
plt.annotate('data = (%.1f, %.1f)'%(xdata,ydata),
            (xdata, ydata), xytext =(-2 * offset, offset),
            textcoords ='offset points',
            bbox = bbox, arrowprops = arrowprops)
```

```python
plt.subplot(2,1,1)
# 1-row,2-column,1-first plot
plt.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
plt.subplot(2,1,2)
plt.plot(x*2,y,'^-k',lw=1,mec='red',mew=4)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
ax2.plot(x*2,y,'^-k',lw=1,mec='red',mew=4)
```

```python
fig,(ax1,ax2)=plt.subplots(2,1)
ax1.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
ax2.plot(x*2,y,'^-k',lw=1,mec='red',mew=4)
```

```python
fig,(ax1,ax2)=plt.subplots(2,1,sharex=True)
ax1.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
ax2.plot(x*2,y,'^-k',lw=1,mec='red',mew=4)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(15,5))
ax1.plot(x,y,'o:m',lw=2,mec='k',mew=4,mfc='pink',)
y_high=np.max(y)
x_y=np.argmax(y)
x_value=x[x_y]
bbox = dict(pad=.5,boxstyle='round' ,fc ="none",edgecolor='black')
ax1.annotate('Local Maxima',xy=(x_value,y_high),xytext=(x_value+5,y_high),arrowprops=dict(
facecolor='pink',shrink=0.005,),bbox=bbox)
ax1.set_title('Annoted')
ax2.plot(x*2,y,'^-k',lw=1,mec='red',mew=4)
```

```python
data={'C':20,'C++':15,'Java':30,'Python':35,'Php':20}
courses=list(data.keys())
values=list(data.values())
fig,ax=plt.subplots(1,1,figsize=(5,10))
ax.bar(courses,values,width=0.4)
```

```python
plt.figure(figsize=(10,5))
plt.bar(courses,values,width=0.7,color='#86AF33')
```

```python
plt.figure(figsize=(10,5))
plt.bar(courses,values,width=0.7,color='#C776AA')
# #C776AA ---hex color code
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(courses,values,width=0.7,color='#C776AA',edgecolor='black',linewidth=2,tick_label=courses,
      alpha=0.5,capstyle='round',hatch='x',error_kw={'capsize':9})
plt.title('Languages And their Market',fontdict={'fontsize':18,'fontstyle':'italic'},color='#86AF33',)
plt.xlabel('Languages',color='#EB5367',loc='center',fontdict={'fontsize':20})
plt.xticks(rotation=30,color='#3CA1A4',fontsize=15)
plt.ylabel('Market',color='#EB5367',fontdict={'fontsize':20})
plt.yticks(rotation=70,color='#3CA1A4',fontsize=15,)
plt.legend(['lan'],loc='best',fontsize='small',title='Market',labelcolor='maroon',shadow=True,edgecolor='black',title_fontsize='large')
plt.grid(True)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(courses,values,color='#C776AA',edgecolor='black',linewidth=2,tick_label=courses,
      alpha=0.5,capstyle='round',hatch='.',error_kw={'capsize':9})
plt.title('Languages And their Market',fontdict={'fontsize':18,'fontstyle':'italic'},color='#86AF33',)
plt.ylabel('Languages',color='#EB5367',loc='center',fontdict={'fontsize':20})
plt.yticks(rotation=30,color='#3CA1A4',fontsize=15)
plt.xlabel('Market',color='#EB5367',fontdict={'fontsize':20})
plt.xticks(rotation=70,color='#3CA1A4',fontsize=15,)
plt.legend(['lan'],loc='best',fontsize='small',title='Market',labelcolor='maroon',shadow=True,edgecolor='black',title_fontsize='large')
plt.grid(True)
plt.text(35,-1.5,'Chaitanyaa',fontsize=15,color='grey',ha='right',va='bottom',alpha=.5)
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='heavy',
             color ='black')
```

```python
plt.subplots(figsize=(12,8))
barwidth=.25
IT=[12,30,1,8,22]
ECE=[28,6,16,5,10]
CSE=[29,3,24,25,17]
br1=np.arange(len(IT))
br2=[x+barwidth for x in br1]
br3=[x+barwidth for x in br2]
plt.bar(br1,IT,color='r',width=barwidth,edgecolor='k',label='IT')
plt.bar(br2,ECE,color='g',width=barwidth,edgecolor='k',label='ECE')
plt.bar(br3,CSE,color='b',width=barwidth,edgecolor='k',label='CSE')
plt.xticks(br2,['2015','2016','2017','2018','2019'],fontsize=20)
plt.legend(title='branch',title_fontsize='x-large',facecolor='pink',edgecolor='black')
plt.title('MultipleBarChartOF Branch',fontdict={'fontsize':25,'color':'maroon'},loc='left')
```

```python
import pandas
df = pandas.DataFrame(dict(graph=['Item one', 'Item two', 'Item three'],
                           n=[3, 5, 2], m=[6, 1, 3])) 

ind = np.arange(len(df))
width = 0.4

fig, ax = plt.subplots()
ax.barh(ind, df.n, width, color='red', label='N')
ax.barh(ind + width, df.m, width, color='green', label='M')

ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width - 1, len(df)])
ax.legend()
```

```python
boys=(20,35,30,35,27)
girls=(25,32,34,20,25)
wd=np.arange(5)
width=.25
boystd=(2,3,4,1,2)
girlsstd=(3,5,2,3,3)
plt.bar(wd,boys,width,yerr=boystd)
plt.bar(wd,girls,width,bottom=boys,yerr=girlsstd)
plt.legend(['boys','girls'],title="Gender")
```

```python
boys=(20,35,30,35,27)
girls=(25,32,34,20,25)
wd=np.arange(5)
width=.25
boystd=(2,3,4,1,2)
girlsstd=(3,5,2,3,3)
plt.figure(figsize=(15,7))
plt.barh(wd,boys,width,xerr=boystd)
plt.barh(wd,girls,width,left=boys,xerr=girlsstd)
plt.legend(['boys','girls'],title="Gender")
plt.yticks([0,1,2,3,4],['F1','F2','F3','F4','F5'],rotation=75)
plt.ylabel('CONTRIBUTION',color='grey',fontdict={'fontsize':15})
plt.title('Boys nd Girls Contribution',loc='center',fontdict={'fontsize':15,'color':'green'})
plt.xlabel('VALUES',color='grey',fontdict={'fontsize':15})
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
plt.pie(data,labels=cars)
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
explode=(.1,0,.2,.3,0,.5)
colors=('orange','pink','brown','grey','indigo','beige')

plt.pie(data,labels=cars,autopct='%1.1f%%',explode=explode,shadow=True,colors=colors,startangle=30,wedgeprops={'linewidth':3,'edgecolor':'black'},
       textprops=dict(color='magenta'))

plt.legend(cars,title='Cars',loc='center right',title_fontsize='x-large')
plt.title('Cutomized Pie Chart',fontdict={'fontsize':20,'color':'black','fontstyle':'italic'})
```
